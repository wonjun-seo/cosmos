This notebook explores how numeric variables relate to each other using correlation techniques. We use both the Titanic dataset and a simulated dataset.

## 1. Correlation Heatmap (Titanic Dataset)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation (numeric columns only)
corr = df_titanic.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Titanic Dataset: Correlation Matrix')
plt.show()
```

###  Interpretation:
- Look for strong positive/negative values to identify linear relationships.
- Variables like `fare` and `pclass` may show negative correlation (higher class  lower fare).
- `sibsp` and `parch` might show mild association due to family structure.
